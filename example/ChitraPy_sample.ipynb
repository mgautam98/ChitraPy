```
from ChitraPy import filters, helpers
import matplotlib.pyplot as plt
import numpy as np
```

```
# Load a sample Image

!wget https://i.imgur.com/D24n5DL.png
img = plt.imread('./D24n5DL.png')
plt.imshow(img)
```

```
# invert
invert = filters.invert(img)
plt.imshow(invert)
```

```
# gray image
gray = helpers.rgb2gray(img)
plt.imshow(gray, cmap = plt.get_cmap('gray'))
```

```
negate = helpers.InvertGrayImg(gray)
plt.imshow(negate, cmap = plt.get_cmap('gray'))
```

```
# right sobel operator
right = filters.right_sobel(img)
plt.imshow(right, cmap = plt.get_cmap('gray'))
```

```
# rotate
rotate = filters.rotate(img)
plt.imshow(rotate)
```
